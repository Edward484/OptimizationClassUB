```python
import QPGC
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import LinearSVC
```

```python
num_points_class1 = 50
num_points_class2 = 50
k = 1
```

```python
class1 = np.array(np.random.multivariate_normal([0.,0.], [[1,-0.9],[-0.9,1]], num_points_class1))
class2 = np.array(np.random.multivariate_normal([3.,6.], [[1,0],[0,1]], num_points_class2))
non_separable = np.array(np.random.multivariate_normal([1.,2.], [[1,0],[0,1]], 500))
```

```python
nd = np.concatenate((class1, class2), axis=0)
labels = np.concatenate((-np.ones(num_points_class1),np.ones(num_points_class2)), axis=0)
```

```python
total_points = num_points_class1+num_points_class2
G = np.diag(labels) @ nd @ np.transpose(nd) @ np.diag(labels)
g = -np.ones(total_points)
A = labels.reshape(total_points,1)
b = np.zeros(1)
C = np.concatenate((np.diag(np.ones(total_points)), np.diag(-np.ones(total_points)))).T
d = np.concatenate((np.zeros(total_points), (-1)*k*np.ones(total_points)))
```

```python
alpha = QPGC.QPGenericConstraints(G,g,A,C,b,d)
```

```python
alpha
```

```python
w = (alpha * labels)[:,None] * nd
w = np.sum(w, axis=0)
print(w)
```

```python
index = np.argmax(alpha)
b = labels[index] -(w.T @ nd[index])
```

```python
plt.scatter(nd[:,0], nd[:,1],c=labels, alpha=0.5)
a = -w[0] / w[1]
xx = np.linspace(-5, 8)
yy = (a * xx) - ((b) / w[1])
yy_max = (a * xx) - ((b+1) / w[1])
yy_min = (a * xx) - ((b-1) / w[1])
plt.plot(xx,yy)
plt.plot(xx,yy_max)
plt.plot(xx,yy_min)
```

```python
# Create support vector classifier
svc = LinearSVC(C=1.0)

# Train model
model = svc.fit(nd, labels)
w = model.coef_[0]
b = model.intercept_

```

```python
plt.scatter(nd[:,0], nd[:,1],c=labels, alpha=0.5)
a = -w[0] / w[1]
xx = np.linspace(-2.5, 2.5)
yy = (a * xx) - ((b) / w[1])
plt.plot(xx,yy)
```

```python
num_points_class1 = 50
num_points_class2 = 50
num_points_nonsep = 50
k = 1
class1 = np.array(np.random.multivariate_normal([0., 0.], [[1, -0.9], [-0.9, 1]], num_points_class1))
class2 = np.array(np.random.multivariate_normal([3., 6.], [[1, 0], [0, 1]], num_points_class2))
non_separable = np.array(np.random.multivariate_normal([1., 2.], [[1, 0], [0, 1]], num_points_nonsep))

nd = np.concatenate((class1, class2), axis=0)
labels = np.concatenate((-np.ones(num_points_class1),np.ones(num_points_class2)), axis=0)
```

```python
def step(w,b, gamma, x, y, lmb):
    if (w.T @ x + b)*y > 1:
        w_new = w - gamma * w * lmb
        b_new = b
    else:
        w_new = w - gamma * (w * lmb - y * x)
        b_new = b + gamma * y
    return w_new, b_new
```

```python
def stoch_grad(epochs, w,b,lmb, gammas,xs,ys):          #gammas -> []*len(epochs)

    randInds = list(range(len(ys)))
    for i in range(epochs):
        np.random.shuffle(randInds)

        for j in randInds:
            w,b = step(w,b,gammas[i],xs[j],ys[j],lmb)
    return w,b
```

```python
w, b = stoch_grad(1000, np.zeros(2), np.zeros(1), 0.01, [0.001] * 1000, nd, labels)
print(w, b)
plt.scatter(nd[:, 0], nd[:, 1], c=labels, alpha=0.5)
a = -w[0] / w[1]
xx = np.linspace(-5, 8)
yy = (a * xx) - ((b) / w[1])
yy_max = (a * xx) - ((b + 1) / w[1])
yy_min = (a * xx) - ((b - 1) / w[1])
plt.plot(xx, yy)
plt.plot(xx, yy_max)
plt.plot(xx, yy_min)
```

```python
def step_mini_batch(w,b, gamma, xs, ys, lmb):
    sum_w = 0
    sum_b = 0
    for i in range(len(xs)):
        if (w.T @ xs[i] + b)*ys[i] > 1:
            sum_w += 0
            sum_b += 0
        else:
            sum_w +=  - ys[i] * xs[i]
            sum_b += -ys[i]
    w_new = w - gamma * (lmb * w + sum_w)
    b_new = b - gamma * sum_b
    return w_new, b_new
```

```python
def stoch_grad_mini_batch(epochs, w,b,lmb, gammas,xs,ys, batch_size=10):          #gammas -> []*len(epochs)

    randInds = list(range(len(ys)))
    for i in range(epochs):
        np.random.shuffle(randInds)

        for j in range(len(xs)//batch_size + 1):
            inds = randInds[j:min(j+batch_size, len(xs))]
            w,b = step_mini_batch(w,b,gammas[i],xs[inds],ys[inds],lmb)
    return w,b
```

```python
w,b = stoch_grad_mini_batch(100,np.zeros(2), np.zeros(1), 0.01, [0.001]*100, nd, labels)
print(w,b)
```

```python
plt.scatter(nd[:,0], nd[:,1],c=labels, alpha=0.5)
a = -w[0] / w[1]
xx = np.linspace(-5, 8)
yy = (a * xx) - ((b) / w[1])
yy_max = (a * xx) - ((b+1) / w[1])
yy_min = (a * xx) - ((b-1) / w[1])
plt.plot(xx,yy)
plt.plot(xx,yy_max)
plt.plot(xx,yy_min)
```
